```python
%matplotlib inline

import pandas as pd
import cv2
from os.path import isfile, isdir, join, pardir
import matplotlib.pyplot as plt
import random
```

```python
result = pd.read_csv('Xception_35+InceptionResNetV2_274+InceptionV3_100.csv')
image_name = result['img']
pred = result.drop(['img'], axis=1)

driver_dataset_folder_path = join(pardir, 'data', 'driver_dataset')
test_image_path = join(driver_dataset_folder_path, 'test', '0')

dict = {"c0":"safe driving",
        "c1":"texting - right", 
        "c2":"talking on the phone - right", 
        "c3":"texting - left", 
        "c4":"talking on the phone - left",
        "c5":"operating the radio",
        "c6":"drinking",
        "c7":"reaching behind",
        "c8":"hair and makeup",
        "c9":"talking to passenger"}

plt.figure(figsize=(12, 24))
for i in range(20):
    plt.subplot(7, 3, i+1)
    image_idx = random.randint(0, 79727)
    img = cv2.imread(join(test_image_path, image_name[image_idx]))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    plt.title('%s | %.2f' % (dict[pred.iloc[image_idx].idxmax()], 
                                     pred.iloc[image_idx].max()))
    plt.axis('off')
    plt.imshow(img)
```
